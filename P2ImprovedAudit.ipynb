```python
#Block 1
#Audit code, does not take time to run, Update AUDIT_LIST for different Tags to audit
import xml.etree.cElementTree as ET
from collections import defaultdict
import re
import pprint
import operator

#OSM file name to be audited, change as needed
OSMFILE = "austin_texas.osm"
#OSMFILE = "austin_texas_sample.osm"

#Usage depricated
AUDIT_CATEGORY = "service"

#changet AUDIT_LIST to reflect the Tags that you wish to Audit then execute "Block 1"
#To find a list of tag names sorted by their frequency, execute "Block 2" then "Block 3"
AUDIT_LIST = ["addr:postcode","addr:state", "cuisine", "religion","addr:city", "shop", "amenity"]


street_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)


expected = ["Street", "Avenue", "Boulevard", "Drive", "Court", "Place", "Square", "Lane", "Road", 
            "Trail", "Parkway", "Commons"]

expname = ["Street", "Avenue", "Boulevard", "Drive", "Court", "Place", "Square", "Lane", "Road", 
            "Trail", "Parkway", "Commons", "Highway", "Circle", "North", "West", "South", "East"]

#Mapping list to aid in cleaning street names
# UPDATE THIS VARIABLE
mapping = { "St": "Street",
            "St.": "Street",
            "Ave": "Avenue",
            "Rd.": 'Road',
            "Ave.": "Avenue",
            "Blvd": "Boulevard",
            "Blvd.": "Boulevard",
            "CR": "Circle",
            "Cir": "Circle",
            "Ct": "Court",
            "Cv": "Cove",
            "Dr": "Drive",
            "Dr.": "Drive",
            "Expwy": "Expressway",
            "Hwy": "Highway",
            "Ln": "Lane",
            "N": "North",
            "Pkwy": "Parkway",
            "RD": "Road",
            "Rd": "Road",
            "W": "West",
            "WEST": "West",
            "lane": "Lane",
            "street": "Street",
            }

#Audits only the last word of the street name
def audit_street_type(street_types, street_name):
    m = street_type_re.search(street_name)
    if m:
        street_type = m.group()
        if street_type not in expected:
            street_types[street_type].add(street_name)

#Audits the entire street name for over-abbreviation and inconsistencies
def audit_street_name(street_names, street_name):
    namesplit = street_name.strip().split(' ')
    for i in namesplit:
        if i not in expname:
            street_names[i].add(street_name)


#Filter tests for particular audited category
def is_street_name(elem):
    return (elem.attrib['k'] == "addr:street")

def is_amenity(elem):
    return (elem.attrib['k'] == AUDIT_CATEGORY)

#Generalized filter for audited category
def is_audit_cat(elem, cat):
    return (elem.attrib['k'] == cat)

#Generalized processor/counter for audited category
def audit_amenity_type(amen_types, amen_name):
    if amen_types.has_key(amen_name):
        amen_types[amen_name] += 1
    else:
        amen_types[amen_name] = 1


def audit(osmfile):
    
    #initializations
    osm_file = open(osmfile, "r")
    street_types = defaultdict(set)
    street_names = defaultdict(set)
    amen_types = {}
    list_of_dict = []
    for l in range(0, len(AUDIT_LIST)):
        list_of_dict.append({})
    tag_attrib = {}
    count = 0
    
    #Loop through elements for node and way
    for event, elem in ET.iterparse(osm_file, events=("start",)):
        
        if elem.tag == "node" or elem.tag == "way":
            count += 1
            for tag in elem.iter("tag"):
                if tag_attrib.has_key(tag.attrib['k']):
                    tag_attrib[tag.attrib['k']] += 1
                else:
                    tag_attrib[tag.attrib['k']] = 1
                
                if is_street_name(tag):
                    audit_street_type(street_types, tag.attrib['v'])
                    audit_street_name(street_names, tag.attrib['v'])
                elif is_amenity(tag):
                    audit_amenity_type(amen_types, tag.attrib['v'])
                else:
                    ndx = 0
                    for i in AUDIT_LIST:
                        if is_audit_cat(tag, i):
                            audit_amenity_type(list_of_dict[ndx], tag.attrib['v'])
                        ndx +=1
                        
    #Output
    print "done... "
    print "Total lines processed: " + str(count)
    return street_types, street_names, tag_attrib, amen_types, list_of_dict, count

#not used part of original code
def update_name(name, mapping):
    
    m = street_type_re.search(name)
    if m:
        if mapping.has_key(m.group()):
            name = name.replace(m.group(),mapping[m.group()])
    return name

#test function, no longer used
def test():
    
    st_types, st_names, tl, amt, lod, count = audit(OSMFILE)
    #st_types, st_names, tl = audit(OSMFILE)
    #assert len(st_types) == 3
    print len(st_types)
    print len(st_names)
    print len(tl)
    #pprint.pprint(dict(st_types))
    #pprint.pprint(dict(st_names))
    
    #for st_type, ways in st_types.iteritems():
    #    for name in ways:
    #        better_name = update_name(name, mapping)
    #        print name, "=>", better_name
    #        if name == "West Lexington St.":
    #            assert better_name == "West Lexington Street"
    #        if name == "Baldwin Rd.":
    #           assert better_name == "Baldwin Road"
    

#if __name__ == '__main__':
    #test()
```
